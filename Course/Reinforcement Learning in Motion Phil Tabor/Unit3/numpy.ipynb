## Unit variance

```python
def rolling_average(average, instance, new_score):
    if instance < 1:
        return average + (instance *(new_score - average))
    return average + (1 / (instance + 1))*(new_score - average)

def test(handle)->None:
    mean = np.zeros((len(handle)))
    for i in range(30):
        for idx, n in enumerate(handle):
            a = np.random.normal(n)
            mean[idx] = rolling_average(mean[idx], i+1, a)
    print (mean, " : ", np.mean(mean))

```

## Convergence

Convergence

```python
import matplotlib.pyplot as plt
import numpy
typical = []
static_alpha = []
squared_alpha = []
for i in range(1000):
    b = i+1
    typical.append(1/b)
    static_alpha.append(b)
    squared_alpha.append(min(b**2, 1000))

plt.plot(range(1, len(typical) + 1), typical, label="typical")
plt.plot(range(1, len(static_alpha) + 1), static_alpha, '.-', label="static_alpha")
plt.plot(range(1, len(squared_alpha) + 1), squared_alpha, '.-', label="squared_alpha")
plt.legend()  # To draw legend
plt.show()
```
